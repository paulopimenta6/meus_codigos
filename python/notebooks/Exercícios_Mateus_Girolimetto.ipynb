Faça uma função que retorne o reverso de um número inteiro informado. Por exemplo: 127 -> 721.

```python
def reverso(n):
    '''
    Essa função tem como objetivo retornar o reverso de um número 
    Válido para números Interios e Positivos
    '''
    n = str(n) # Criando variável que atribua o valor do parâmetro da função e convertendo ela em string
    n_reverso = int(n[::-1]) # Sintaxe string[::] usada para fatiar uma string; s[começo(incluso):fim (não incluso): passos]
    p = print(f'O reverso do número {n} é {n_reverso} ') 
    return p # A função retornará p
```

```python
n = input('Digite um número:')
reverso(n)
```

Faça uma função para converter, que dado um valor numérico de temperatura e um valor de caractere para a escala (Celsius ou Farenheit) converta o número para a escala desejada e retorna o valor.

```python
def conversão(e):
    '''
    Função para converter temperaturas entre as escalar °C e °F 
    '''
    if e == 'c' or e =='C':
        n = round(float(input('Digite o valor da temperatura em °F:')),2)
        t_c = round(5/9*(n - 32),2)
        p = print(f'A temperatura {n}°F equivale a {t_c}°C')
        return p 
    elif  e == 'f' or e =='F':
        n = round(float(input('Digite o valor da temperatura em °C:')))
        t_f =  round(9/5*n + 32,2)
        p = print(f'A temperatura {n}°C equivale a {t_f}°F')
        return p
    else:
        p = print('Alguma informação errada. Comece novamente')
        return p
```

```python
def horário(h,m):
    '''
    Função que convete notação de horas
    '''
    if h <=12 and m < 60:
        p = print(f'O horário é {h}:{m} A.M')
        return p
    elif h > 12 and h < 24 and m < 60:
        p = print(f'O horário é {h - 12}:{m} P.M')
        return p
    else:
        p = print('Alguma informação errada. Comece novamente')
        return p 
```

# Questão 3

Faça uma função que recebe dois parâmetros, uma matriz numpy e uma constante, e utiliza estruturas de repetição para multiplicar cada elemento da matriz e retorna a matriz multiplicada. 

```python
def função(m, c):
    lin,col = m.shape #Saber a dimensão da matriz 
    m_m = np.ones((lin, col), dtype=m.dtype) # Matriz que irá armazenar os dados
    for i in range(lin):
        for j in range(col):
            m_m[i,j] = m[i,j]*c 
    p = print(f'\nA matriz original é: \n{m}\n e a matriz multiplicada pela constante {c} é: \n{m_m}')
    return p

lista = [] # Criando lista vazia para adicionar os valores
linha = [float(i) for i in input('Digite os valores da linha (separados por espaço):').split(' ')] # Pedindo usuário para inserir valores
lista.append(linha) # Adiciona os valores a lista antes criada
msg = input('\nDeseja acrescentar outra linha?: s/n ')
while msg == 's': # Estrutura de repetição para ir adicionando elementos na matriz
    linha = [float(i) for i in input('\nDigite os valores da linha (separados por espaço):').split(' ')]
    lista.append(linha)
    msg = input('\nDeseja acrescentar outra linha?: s/n ')

m = np.array(lista)
c = float(input('\nDigite o valor da constante:'))

função(m,c)
```

## Modelo de Regressão

### Conceito

O modelo de regressão é uma técnica/ferramenta estatística que tem por finalidade identificar e quantificar a relação entre variáveis. Tais variáveis são conhecidas como variável dependente (ou resposta) e  variável independente (ou preditoras). O principal objetivo é prever o valor da variável dependente baseado no valor de uma ou mais variáveis independente. 

### Tipos de Modelos de Regressão

- Regressão Linear Simples:

  Analisa a relação linear entre uma variável dependente e uma variável independente. A equação geral é:
  \begin{equation}
y = \beta_0 + \beta_1 x + \varepsilon
\end{equation}

- Regressão Linear Múltipla:

  Estende a regressão linear simples para incluir múltiplas variáveis independentes.A equação geral é:
 \begin{equation}
y = \beta_0 + \sum_{i=1}^{n} \beta_i x_i + \varepsilon
\end{equation}

- Regressão Logística (Usado nos modelos de classifição):

  Utilizada quando a variável dependente é categórica, especialmente para modelos binários (por exemplo, sim/não, sucesso/falha). A equação geral é:
  \begin{equation}
\log\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
\end{equation}

- Regressão Polinomial:

  Captura relações não lineares ao incluir termos polinomiais das variáveis independentes. A equação geral é:
  \begin{equation}
y = \beta_0 + \beta_1 x + \beta_2 x^2 + \dots + \beta_n x^n + \varepsilon
\end{equation}

- Regressão Ridge e Lasso:

  Técnicas que incorporam regularização para lidar com multicolinearidade e overfitting. A equações são:

  A equação da regressão Ridge minimiza a seguinte função de custo:
  \begin{equation}
\min_{\beta} \left( \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} \beta_j^2 \right)
\end{equation}

  A equação da regressão Lasso minimiza a seguinte função de custo:
\begin{equation}
\min_{\beta} \left( \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} |\beta_j| \right)
\end{equation}

- Regressão de Poisson:

    Aplicada em casos onde a variável dependente é uma contagem ou taxa de ocorrência. A equação geral é:

```python
# Avaliação do modelo 

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = model.score(X_test, y_test)

# Exibindo resultados
print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

# Resposta

## Modelo de Regressão

### Conceito

Modelos de classificação fazem parte do aprendizado de máquina supervisionado. Assim como nos modelos de Regressão Linear, eles são usados para previsão. Baseado em um conjunto de características (ou features) é possível criar modelos para prever rótulos categóricos (ou classes). Vale ressaltar que as classes são variáveis discretas

### Principais Tipos de Classificação

- Classificação Binária:

  Quando há apenas duas classes possíveis. Exemplo: Classificar presença de doença (Doente = SIM/1 ; Não Doente = NÃO/0).

- Classificação Multiclasse:

    Quando há mais de duas classes. Exemplo: Identificar a espécie de uma planta/flor com três categorias possíveis (Setosa, Versicolor, Virginica)

- Classificação Multirrótulo:

  Quando cada exemplo pode pertencer a mais de uma classe simultaneamente. Exemplo: Um filme pode ser classificado como ação, aventura e romance ao mesmo tempo.


### Principais Algoritmos de Classificação

- Regressão Logística:

  Bom para problemas binários.
  Utiliza uma função sigmoide para prever a probabilidade de uma classe.

- K-Nearest Neighbors (KNN):

    Classifica um novo ponto com base nos "K" pontos de dados mais próximos.

- Árvores de Decisão:

    Cria uma árvore de decisões baseada em condições para separar as classes.

- Random Forest:

    Um conjunto (ensemble) de várias árvores de decisão.
    Usa votação para determinar a classe final.
    Mais robusto do que uma única árvore de decisão.

- Support Vector Machine (SVM):

    Tenta encontrar a linha (ou hiperplano) que separa melhor as classes.
    Útil para problemas onde as classes são bem separáveis.

- Naive Bayes:

    Baseado no Teorema de Bayes, assume que as características são independentes entre si

- Redes Neurais:

    São modelos poderosos que tentam imitar a forma como o cérebro humano processa informações.
    Útil para tarefas complexas, como reconhecimento de imagem e processamento de linguagem natural.

### Como Avaliar Modelos de Classificação 

- Acurácia:

  Percentual de previsões corretas.
  Funciona bem em conjuntos de dados balanceados, mas pode ser enganosa quando as classes são desbalanceadas.

- Matriz de Confusão:

  Uma tabela que mostra a contagem de previsões corretas e incorretas para cada classe.

# Resposta

## Conceito

Analises de Componentes Principais (PCA) é um método estatístico usado para redução de dimensões.

O métedo reduz a quantidade de vaiáveis/características do banco de dados enquanto mantém as informações mais importantes para a análise.

Isso facilita a visualização dos dados, reduz custos computacionais e ajuda a prevenir overfitting quando modelos de ML são construídos 

## Conceitos Chave

- Componentes Principais
  São os novos eixos formados a partir da combinação linear das características originais, onde cada eixo aponta para a direção de máxima variância

- Variância Explicada
  Os autovalores correspondem à variância explicada por cada componente principal. Examinando a razão de variância explicada podemos decidir quantos componentes manter

## Como interpretar os resultados do PCA

- Componentes Principais
  A matriz de saída mostra os dados tranformados no novo espaço definido pelos componentes principais

- Razão de Variancia Explicada
  Mostra o quanto to total da variância dos dados é capturado por cada componente principal



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
```

```python
# Importando dados
df = pd.read_csv('penguins.csv')
df.head()
```

```python
# Informações dos dados
df.info()
```

```python
# Checando valores nulos
df.isnull().sum()
```

```python
# Excluindo valores nulos
df.dropna(inplace= True)

df.isnull().sum()
```

```python
# Conferindo valores da categoria 'sex'
df['sex'].value_counts()
```

```python
# Eliminando valor desnecessário 
df = df[df['sex']!= '.']
df['sex'].value_counts()
```

```python
# Conferindo e Excluindo Outliers

# Criando Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(df,palette='Set2')
```

```python
df = df[(df['flipper_length_mm'] < 4000) & (df['flipper_length_mm'] >= 0) ]

plt.figure(figsize=(10, 6))
sns.boxplot(df,palette='Set2')
```

```python
# Transformando variáveis categóricas em numéricas

label_encoder = LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df.head()
```

```python
# Padronizando valores

scaler = StandardScaler()
X = scaler.fit_transform(df)

df_preprocessed = pd.DataFrame(X, columns=df.columns)
df_preprocessed.head()
```

```python
# Criando PCA
pca = PCA(n_components=None)
df_pca = pca.fit(df_preprocessed)
df_pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=2)
df_PCA = pca.fit_transform(df_preprocessed)

df_PCA = pd.DataFrame(df_PCA,columns=['PC1','PC2'])
df_PCA.head()
```

```python
# Criando KNN
inertia = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,random_state=42).fit(df_PCA)
    inertia.append(kmeans.inertia_)
```

```python
# Gráfico Modelos dos Cotovelos

plt.plot(range(1,11),inertia, marker = 'o')
plt.xlabel('Numero de Classes')
plt.ylabel('Inercia')
plt.title('Método dos Cotovelos')
plt.show()
```

```python
# Gráfico K-means

n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(df_PCA)
plt.scatter(df_PCA.iloc[:, 0], df_PCA.iloc[:, 1], c=kmeans.labels_, cmap="viridis")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title(f"K-means Clustering (K={n_clusters})")
plt.show()
```
