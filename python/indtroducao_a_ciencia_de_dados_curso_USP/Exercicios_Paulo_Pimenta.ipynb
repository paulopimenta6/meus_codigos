## **Exercício 2**

```python
def converter_temperatura(valor, escala):
    if escala.upper() == 'C':
        # Converte de Fahrenheit para Celsius
        return (valor - 32) * 5.0 / 9.0
    elif escala.upper() == 'F':
        # Converte de Celsius para Fahrenheit
        return (valor * 9.0 / 5.0) + 32
    else:
        raise ValueError("Escala inválida. Use 'C' para Celsius ou 'F' para Fahrenheit.")

# Solicitando a entrada do usuário
try:
    temperatura = float(input("Informe o valor da temperatura: "))
    escala_desejada = input("Informe a escala para conversão ('C' para Celsius, 'F' para Fahrenheit): ")

    # Determinando a unidade de destino
    unidade_destino = 'C' if escala_desejada.upper() == 'F' else 'F'

    # Realizando a conversão
    temperatura_convertida = converter_temperatura(temperatura, escala_desejada)

    # Imprimindo o resultado
    print(f"{temperatura}°{escala_desejada.upper()} -> {temperatura_convertida:.2f}°{unidade_destino}")

except ValueError as e:
    print(f"Erro: {e}")
```

## **Exercício 3**

```python
def converter_hora_24_para_12(hora, minuto):
    """
    Converte uma hora no formato de 24 horas para o formato de 12 horas.

    Parâmetros:
    - hora (int): Hora no formato de 24 horas (0 a 23).
    - minuto (int): Minutos (0 a 59).

    Retorna:
    - str: Hora no formato de 12 horas com AM/PM.
    """
    if not (0 <= hora <= 23) or not (0 <= minuto <= 59):
        raise ValueError("Hora deve estar entre 0 e 23 e minuto entre 0 e 59.")

    periodo = "A.M." if hora < 12 else "P.M."

    # Ajusta a hora para o formato de 12 horas
    hora_12 = 12 if hora % 12 == 0 else hora % 12  # Converte 0 para 12

    return f"{hora_12}:{minuto:02d} {periodo}"

# Solicitando entrada do usuário
try:
    hora_input = int(input("Informe a hora (0-23): "))
    minuto_input = int(input("Informe os minutos (0-59): "))

    # Realizando a conversão
    hora_convertida = converter_hora_24_para_12(hora_input, minuto_input)

    # Imprimindo o resultado
    print(f"A hora convertida é: {hora_convertida}")

except ValueError as e:
    print(f"Erro: {e}")
```

## **Exercício 3**

```python
import numpy as np

def multiplicar_por_constante(matriz, constante):
    # Cria uma matriz de saída com o mesmo formato da matriz original
    matriz_saida = np.zeros(matriz.shape)

    # Laços para percorrer cada elemento da matriz
    for linha in range(matriz.shape[0]):  # Itera sobre as linhas
        for coluna in range(matriz.shape[1]):  # Itera sobre as colunas
            matriz_saida[linha][coluna] = matriz[linha][coluna] * constante  # Multiplica o elemento pela constante

    return matriz_saida

# Solicita o número de linhas e colunas da matriz
total_linhas = int(input("Digite o número de linhas da matriz: "))
total_colunas = int(input("Digite o número de colunas da matriz: "))

# Lista para armazenar as linhas da matriz
dados_matriz = []

# Coleta os elementos da matriz
print(f"\nDigite os elementos da matriz {total_linhas}x{total_colunas}, separados por espaço:")
for linha in range(total_linhas):
    elementos_linha = input(f"Linha {linha + 1}: ").strip().split()
    dados_matriz.append([int(elemento) for elemento in elementos_linha])  # Converte para inteiros

# Converte a lista de listas em uma matriz NumPy
matriz_inicial = np.array(dados_matriz)

# Solicita o valor da constante
fator_multiplicacao = int(input("\nDigite a constante para multiplicação: "))

# Chama a função
matriz_resultante = multiplicar_por_constante(matriz_inicial, fator_multiplicacao)

# Exibe a matriz original e a matriz resultante
print("\nMatriz Original:")
print(matriz_inicial)

print(f"\nMatriz após multiplicação por {fator_multiplicacao}:")
print(matriz_resultante)
```

## Modelo de Regressão

### Definição

O modelo de regressão é uma abordagem estatística que visa identificar e quantificar a relação entre diferentes variáveis. Essas variáveis são classificadas como variável dependente (ou resposta) e variável independente (ou preditoras). O objetivo principal é prever o valor da variável dependente com base nos valores de uma ou mais variáveis independentes.

### Tipos de Modelos de Regressão

- **Regressão Linear Simples:**

  Esta técnica analisa a relação linear entre uma variável dependente e uma única variável independente. A equação geral é expressa como:
  
  \begin{equation}
  y = \beta_0 + \beta_1 x + \varepsilon
  \end{equation}

- **Regressão Linear Múltipla:**

  Esta abordagem estende a regressão linear simples para incluir várias variáveis independentes. A equação geral é dada por:
  
  \begin{equation}
  y = \beta_0 + \sum_{i=1}^{n} \beta_i x_i + \varepsilon
  \end{equation}

- **Regressão Logística (Usada em Modelos de Classificação):**

  Este modelo é utilizado quando a variável dependente é categórica, especialmente em casos binários (como sim/não ou sucesso/falha). A equação geral é:
    
  \begin{equation}
  \log\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
  \end{equation}

- **Regressão Polinomial:**

  Essa técnica capta relações não lineares ao incluir termos polinomiais das variáveis independentes. A equação geral é expressa como: 
  
  \begin{equation}
  y = \beta_0 + \beta_1 x + \beta_2 x^2 + \dots + \beta_n x^n + \varepsilon
  \end{equation}

- **Regressão Ridge e Lasso:**

  Essas técnicas incorporam regularização para abordar problemas de multicolinearidade e overfitting. As equações são:

  A regressão Ridge minimiza a seguinte função de custo:  
  
  \begin{equation}
  \min_{\beta} \left( \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} \beta_j^2 \right)
  \end{equation}

  A regressão Lasso minimiza a seguinte função de custo:
    
  \begin{equation}
  \min_{\beta} \left( \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} |\beta_j| \right)
  \end{equation}

- **Regressão de Poisson:**

  Essa abordagem é aplicada em situações onde a variável dependente representa uma contagem ou taxa de ocorrência. A equação geral é:
    
  \begin{equation}
  \log(\lambda) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
  \end{equation}

# 8 - Classificação

# Modelo de classificação

## Conceito

Modelos de classificação fazem parte do aprendizado de máquina supervisionado. Assim como nos modelos de Regressão Linear, eles são usados para previsão. Baseado em um conjunto de características (ou features), é possível criar modelos para prever rótulos categóricos (ou classes). Vale ressaltar que as classes são variáveis discretas.

## Principais Tipos de Classificação

- **Classificação Binária**:  
  Quando há apenas duas classes possíveis. Exemplo: Classificar presença de doença (Doente = SIM/1; Não Doente = NÃO/0).

- **Classificação Multiclasse**:  
  Quando há mais de duas classes. Exemplo: Identificar a espécie de uma planta/flor com três categorias possíveis (Setosa, Versicolor, Virginica).

- **Classificação Multirrótulo**:  
  Quando cada exemplo pode pertencer a mais de uma classe simultaneamente. Exemplo: Um filme pode ser classificado como ação, aventura e romance ao mesmo tempo.

## Principais Algoritmos de Classificação

- **Regressão Logística**:  
  Bom para problemas binários. Utiliza uma função sigmoide para prever a probabilidade de uma classe.

- **K-Nearest Neighbors (KNN)**:  
  Classifica um novo ponto com base nos "K" pontos de dados mais próximos.

- **Árvores de Decisão**:  
  Cria uma árvore de decisões baseada em condições para separar as classes.

- **Random Forest**:  
  Um conjunto (ensemble) de várias árvores de decisão. Usa votação para determinar a classe final. Mais robusto do que uma única árvore de decisão.

- **Support Vector Machine (SVM)**:  
  Tenta encontrar a linha (ou hiperplano) que separa melhor as classes. Útil para problemas onde as classes são bem separáveis.

- **Naive Bayes**:  
  Baseado no Teorema de Bayes, assume que as características são independentes entre si.

- **Redes Neurais**:  
  São modelos poderosos que tentam imitar a forma como o cérebro humano processa informações. Útil para tarefas complexas, como reconhecimento de imagem e processamento de linguagem natural.

## Como Avaliar Modelos de Classificação 

- **Acurácia**:  
  Percentual de previsões corretas. Funciona bem em conjuntos de dados balanceados, mas pode ser enganosa quando as classes são desbalanceadas.

- **Matriz de Confusão**:  
  Uma tabela que mostra a contagem de previsões corretas e incorretas para cada classe.

# 9 - Agrupamento

## Conceito

A Análise de Componentes Principais (PCA) é uma técnica estatística utilizada para reduzir a dimensionalidade dos dados. 

Esse método diminui o número de variáveis ou características em um conjunto de dados, preservando as informações mais relevantes para a análise.

Isso não apenas facilita a visualização dos dados, mas também diminui os custos computacionais e ajuda a evitar o overfitting durante a construção de modelos de aprendizado de máquina.

## Conceitos Chave

- **Componentes Principais**  
  Representam os novos eixos gerados a partir de combinações lineares das variáveis originais, com cada eixo apontando na direção de maior variância.

- **Variância Explicada**  
  Os autovalores estão associados à variância explicada por cada componente principal. Avaliando a proporção da variância explicada, é possível determinar quantos componentes devem ser retidos.

## Como interpretar os resultados do PCA

- **Componentes Principais**  
  A matriz resultante exibe os dados transformados no novo espaço definido pelos componentes principais.

- **Proporção de Variância Explicada**  
  Indica a quantidade total da variância dos dados que é capturada por cada componente principal.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Carregando o conjunto de dados
data = pd.read_csv('BostonHousing.csv')
```

```python
# Visualizando as primeiras linhas do conjunto de dados
print("Primeiras linhas do conjunto de dados:")
print(data.head())
```

```python
# Definindo as variáveis independentes e a variável alvo
X = data.drop(columns=['medv'])  # Remove a variável alvo 'medv'
y = data['medv']  # Guarda a variável alvo para uso posterior, se necessário
```

```python
# Padronizando os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Aplicando PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Variância explicada por cada componente
variancia_explicada = pca.explained_variance_ratio_
print("\nVariância Explicada por Cada Componente:")
for i, var in enumerate(variancia_explicada):
    print(f"Componente {i + 1}: {var:.2f}")
```

```python
# Gráfico da variância explicada
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(variancia_explicada) + 1), variancia_explicada, alpha=0.6, align='center', color='blue')
plt.ylabel('Proporção da Variância Explicada')
plt.xlabel('Componentes Principais')
plt.title('Variância Explicada por Cada Componente Principal')
plt.xticks(range(1, len(variancia_explicada) + 1))
plt.grid(axis='y')
plt.show()
```

```python
# Criando categorias de preço para visualização
data['price_category'] = pd.cut(data['medv'], bins=[0, 15, 25, float("inf")], labels=['baixo', 'médio', 'alto'])
```

```python
# Gráfico dos scores do PCA
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data['price_category'].cat.codes, cmap='viridis', alpha=0.6)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Scores do PCA Coloridos por Categoria de Preço')
plt.colorbar(scatter, ticks=[0, 1, 2], label='Categoria de Preço')
plt.grid()
plt.show()
```
